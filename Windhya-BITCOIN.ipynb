```python
def impute_total(series):
    return series.fillna(series.median())
```

```python
df['btc_total_bitcoins']=df['btc_total_bitcoins'].transform(impute_total)
```

```python
df['btc_trade_volume']=df['btc_trade_volume'].transform(impute_total)
df['btc_blocks_size']=df['btc_blocks_size'].transform(impute_total)
df['btc_median_confirmation_time']=df['btc_median_confirmation_time'].transform(impute_total)
df['btc_difficulty']=df['btc_difficulty'].transform(impute_total)
df['btc_transaction_fees']=df['btc_transaction_fees'].transform(impute_total)
```

```python
df.isnull().sum()
```
